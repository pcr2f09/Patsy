```python
sql_query = """
select *
from actor
where last_name like '%%gen%%'
"""
joined = pd.read_sql_query(sql_query, engine)
joined.head()
```

```python
sql_query = """
select *
from actor
where last_name like '%%li%%'
order by
last_name, first_name
"""
joined = pd.read_sql_query(sql_query, engine)
joined.head()
```

```python
sql_query = """
select country_id, country
from country 
where country IN ('Afghanistan', 'Bangladesh', 'China')
"""
joined = pd.read_sql_query(sql_query, engine)
joined.head()
```

```python
sql_query = """
SELECT last_name, COUNT(*) AS TotalCount
FROM actor
GROUP BY last_name
ORDER BY COUNT(*) DESC
"""
joined = pd.read_sql_query(sql_query, engine)
joined.head()
```
